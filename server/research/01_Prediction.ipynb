```python
def detect_dog(frame):
  detections = model(frame)[0]
  detections_ = []
  for detection in detections.boxes.data.tolist():
    #print(detection)
    x1,y1,x2,y2,score, class_id = detection
    if int(class_id) == 16:
      detections_.append([x1,y1,x2,y2,score])
  return detections_


def crop_only_bb(image_path):
  dog = cv2.imread(str(os.path.join(image_path)))
  (x_center,y_center,bb_height,bb_width,score) = detect_dog(dog)[0]
  image_width = dog.shape[1]
  image_height = dog.shape[0]
  dog_cropped = dog[int(y_center):int(bb_width), int(x_center):int(bb_height),:]
  cv2.imwrite('artifacts/predict/cropped_img.jpg', dog_cropped)
```

```python

crop_only_bb('artifacts/predict/Chihuahua_test2.png')
image = cv2.imread(str(os.path.join('artifacts/predict/cropped_img.jpg')))
image_resized = cv2.resize(image,(img_height,img_width))
image = np.expand_dims(image_resized,axis=0)
pred = PredictModel.predict(image)
pred_class = class_names[np.argmax(pred)]
print(pred_class[10:])
```
